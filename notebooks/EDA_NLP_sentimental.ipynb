```python
import re
import contractions
from unidecode import unidecode

def normalize_text(text):

    text = text.lower()
    text = unidecode(text)  # elimina acentos
    text = re.sub(r"\'scuse", " excuse ", text)
    text = contractions.fix(text) # p.ej can´t -> cannot

    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text) # reemplaza múltiples espacios por uno solo (incluye \t, \n etc)
    text = text.strip()  # elimina espacios al inicio y final

    return text
```

```python
import emoji

def clean_noise(text):
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE) # elimina URLs
    text = re.sub(r'@\w+', '', text) # elimina menciones del usuario @pepe
    text = emoji.demojize(text) # reemplaza emojis por su nombre

    return text
```

```python
def clean_normalize(text):
  text = clean_noise(text)
  text = normalize_text(text)
  return text
```

```python
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
%%capture
# Cargar modelo de spaCy
nlp = spacy.load("en_core_web_sm")

# Descargar stopwords
nltk.download('punkt_tab')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stop_words.discard('not')  # quitamos 'not' de stopwords
stop_words.discard("no")   # opcional
```

```python
def tokenize_text(text):
    return word_tokenize(text)
```

```python
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words and len(word) > 1]
```

```python
def todo(text):
  text = clean_noise(text)
  text = normalize_text(text)
  text = tokenize_text(text)
  text = remove_stopwords(text)
  return text
```

```python
import nltk
from nltk.corpus import words

# Descargar vocabulario estándar (inglés)
nltk.download('words')
vocab = set(words.words())

def detectar_oov(texto):
    tokens = set(texto.lower().split())
    oov_words = tokens.difference(vocab)
    tiene_oov = len(oov_words) > 0
    return tiene_oov, oov_words

# Aplicar al dataframe
df['has_oov'], df['oov_words'] = zip(*df['Text'].apply(detectar_oov))

# Mostrar algunos comentarios que contienen palabras fuera de vocabulario
ejemplos_oov = df[df['has_oov']][['Text', 'oov_words']].head(10)
for i, row in ejemplos_oov.iterrows():
    print(f"Comentario #{i}:\n{row['Text']}\nPalabras OOV: {row['oov_words']}\n")
```
